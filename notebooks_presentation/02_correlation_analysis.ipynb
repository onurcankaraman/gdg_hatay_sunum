```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data_path = '/Users/onurcan/Documents/GDG_HATAY/Dataset.xlsx'
df = pd.read_excel(data_path)
df_clean = df.dropna()
numeric_df = df_clean.select_dtypes(include=[np.number])
corr = numeric_df.corr()
display(corr.round(3))
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', center=0, square=True, linewidths=0.5, cbar_kws={'label':'Korelasyon'})
plt.title('Korelasyon Isı Haritası')
plt.tight_layout()
plt.show()
```

```python
corr_pairs = corr.unstack()
corr_pairs = corr_pairs[corr_pairs != 1.0]
corr_pairs = corr_pairs.sort_values(ascending=False)
print('En yüksek pozitif korelasyonlar:')
print(corr_pairs.head(10).round(3))
print('
En yüksek negatif korelasyonlar:')
print(corr_pairs.tail(10).round(3))
```
